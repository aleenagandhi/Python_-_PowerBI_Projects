```python
import numpy as np
import pandas as pd
movies_df = pd.read_csv('movie.csv')
movies_df.info()
```

```python
movies_df1=movies_df[["movie_title","genres","director_name"]]
movies_df1
```

```python
movies_df1.isnull().sum()
## no null entries
movies_df1.duplicated().sum()
movies_df1

movies_df1 = movies_df1[movies_df1['director_name'] != '-']
```

```python
movies_df1.dropna(inplace=True)
## if there happen to be any null values
```

```python
movies_df1
```

```python
movies_df1["tags"]= movies_df1["movie_title"] +' '+  movies_df1["genres"]+' '+movies_df1["director_name"]
```

```python
movies_df1
```

```python
import re
def replace_special_characters(text):
    return re.sub(r'[^a-zA-Z0-9 ]', ' ', text)

# Apply the function to create a new column
movies_df1['cleaned_tags'] = movies_df1['tags'].apply(replace_special_characters)

movies_df1
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2))
tfidf_matrix1 = vectorizer.fit_transform(movies_df1["cleaned_tags"])
tfidf_matrix1
```

```python
def cleaned_tags(movie_title):
    return replace_special_characters(movie_title.lower())

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(movie_title):
    movie_title = cleaned_tags(movie_title)
    query_vec = vectorizer.transform([movie_title])
    similarity = cosine_similarity(query_vec, tfidf_matrix1).flatten()
    indices = np.argpartition(similarity, -5)[-5:]
    results = movies_df1.iloc[indices].iloc[::-1]
    
    return results
```

```python
results = search("Justice League ")
print("Search results:")
print(results[["movie_title", "genres", "director_name"]])
```

```python
import ipywidgets as widgets
from IPython.display import display

movie_input = widgets.Text(
    value='Gabriela',
    description='Movie Title:',
    disabled=False
)
movie_list = widgets.Output()

def on_type(data):
    with movie_list:
        movie_list.clear_output()
        title = data["new"]
        if len(title) > 5:
            display(search(title))

movie_input.observe(on_type, names='value')


display(movie_input,movie_list)
```
